```python
year = 2025
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
n = 6
fact = 1
for i in range(2, n + 1):
    fact *= i
print("n:", n)
print("factorial:", fact)
```

```python
# Fibonacci
n=10
a=0
b=1
if n > 0:
    for i in range(n):
        print(f"{a}", end=" ")
        sum =a+b
        a=b
        b=sum
```

```python
# Prime Check
n = 29
if n < 2:
    is_prime = False
else:
    is_prime = True
    for d in range(2, n):
        if n % d == 0:
            is_prime = False
            break
if(is_prime):
    print(n ," is a prime number")
else:
    print("Not Prime")
```

```python
# Sum of Digits
n = 1205
s = str(n)
total = 0
for ch in s:
    total += int(ch)
print("n:", n)
print("sum_of_digits:", total)
```

```python
# Reverse a string
text = "Machine"
rev = ""
text[: :-1]
for i in range(len(text)-1,-1,-1):
    rev += text[i]
print("text:", text)
print("reversed:", rev)
```

```python
# Palindrome Check
text="level"
palindrome= text[: :-1]
if (text == palindrome):
    print("Palindrome")
else:
    print("Not Palindrome")
```
