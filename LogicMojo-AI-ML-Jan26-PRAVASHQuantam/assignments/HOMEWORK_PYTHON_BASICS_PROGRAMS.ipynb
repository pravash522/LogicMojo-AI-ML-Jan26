## Program 28: Second Largest (Distinct)

**Problem**: Read `N` and a list of integers. Print the second largest **distinct** number.

**If second largest does not exist**, print: `Not Possible`.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- Second largest distinct value or `Not Possible`

**Sample Input**:
```
5
5 1 5 3 2
```
**Sample Output**:
```
3
```


````python
n= int(input("Enter a number"))
if n > 0:
    if n == 1:
        print("Only 1 element present in the list - Not Possible")
    else:
        nums = []
        for i in range(n):
            nums.append(int(input()))
        firstLargest = nums[0]
        secondLargest = nums[1]
        for i in range(len(nums)):
            if nums[i] > firstLargest:
                secondLargest = firstLargest
                firstLargest = nums[i]
            elif nums[i] > secondLargest and nums[i] != firstLargest:
                secondLargest =nums[i]
        if firstLargest == secondLargest:
            print("Not Possible")
        else:
            print(f"Second largest number is: {secondLargest}")
else:
    print("Enter a value greater than 0")
````

## Program 29: Rotate List Right by K

**Problem**: Read `N`, list of `N` integers, and `K`. Rotate the list to the right by `K` positions.

**Input**:
- Integer `N`
- Next line: `N` integers
- Integer `K`

**Output**:
- Rotated list

**Sample Input**:
```
5
1 2 3 4 5
2
```
**Sample Output**:
```
4 5 1 2 3
```


````python
n= int(input("Enter a number"))
if n > 0:
    nums = []
    for i in range(n):
        nums.append(int(input()))
    k = int(input("Enter k position to rotate"))
    if k > n:
        print("Invalid k: Rotation positions cannot exceed array length")
    else:
        def swapNumbers(nums, start, end):
            while start <= end:
                a= nums[start]
                nums[start] = nums[end]
                nums[end] = a
                start = start+1
                end = end-1
        swapNumbers(nums,0,n-1)
        swapNumbers(nums,0,k-1)
        swapNumbers(nums,k,n-1)
        print(*nums)
else:
    print("Enter value greater than 0")
````
