## 28. Second Largest (Distinct)

```python
n= int(input("Enter n"))
nums = []
for i in range(n):
    nums.append(int(input()))
firstLargest = nums[0]
secondLargest = nums[1]
for i in range(len(nums)):
    if nums[i] > firstLargest:
        secondLargest = firstLargest
        firstLargest = nums[i]
    elif nums[i] > secondLargest and nums[i] != firstLargest:
        secondLargest =nums[i]
print(secondLargest)
```

## 29.  Rotate List Right by K

```python
n= int(input("Enter n"))
nums = []
for i in range(n):
    nums.append(int(input()))
k = int(input("Enter k"))
def swapNumbers(nums, start, end):
    while start <= end:
        a= nums[start]
        nums[start] = nums[end]
        nums[end] = a
        start = start+1
        end = end-1
swapNumbers(nums,0,n-1)
swapNumbers(nums,0,k-1)
swapNumbers(nums,k,n-1)
print(*nums)
```
